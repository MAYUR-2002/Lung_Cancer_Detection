```python
column_change={'GENDER':{'M':1,'F':0},'LUNG_CANCER':{'YES':1,'NO':0}}
```

```python
df=df.replace(column_change)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# MAchine Learning MODel Building
```

```python
# Declaring Feature and Target Varaible
X=df.drop(['LUNG_CANCER','LUNG_CANCER_CAT'],axis=1)
Y=df['LUNG_CANCER']
```

```python
# Split data into training and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Perform Feature Engineering
# Feature Engineering is a process in which we transform raw data into useful that helps us understand our model better and increase
# it's predictive power
```

```python
X_train.describe()
```

```python
X_test.describe()
```

```python
# Feature Scaling
# Used to map all feature variables onto same sacle
# Use MinMaxScaler for normalization
# Use StandardScaler for standardization
```

```python
# Model training
from sklearn.linear_model import LogisticRegression
# Linear Regression model
# Instantiate the model
logreg=LogisticRegression(solver='liblinear',random_state=0)
# fit the model
logreg.fit(X_train,Y_train)
```

```python
# Predict Results
logreg.predict(X_test)
```

```python
# Model is imbalanced
```

```python
# .predict_proba gives the probabilities for target varaiable(0 and 1) in this case in array form
logreg.predict_proba(X_test)[:,1]
```

```python
# We need to balance model using logreg1=LogisticRegression(class_weight='balanced',random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
# Linear Regression model
# Instantiate the model
logreg1=LogisticRegression(class_weight='balanced',random_state=0)
# fit the model
logreg1.fit(X_train,Y_train)
```

```python
Y_pred_test=logreg1.predict(X_test)
```

```python
logreg1.predict_proba(X_test)[:,1]
```

```python
# Check Accuracy Score 
from sklearn.metrics import accuracy_score
print('Model Accuracy Score:',accuracy_score(Y_test,Y_pred_test))
```

```python
# Compare train and test set accuracy
Y_pred_train=logreg1.predict(X_train)
Y_pred_train
```

```python
print("training set accuracy score:",accuracy_score(Y_train,Y_pred_train))
```

```python
# Check for over-fitting or under-fitting
print("training test score : ",logreg1.score(X_train,Y_train))
print("testing set score : ",logreg1.score(X_test,Y_test))
```
